```python
# pandas is a library which has multiple functions will be loaded in the memory
import pandas as pd # from now on instead of typing pandas we can use pd in future
```

```python
dataset=pd.read_csv("Salary_Data.csv") # dataset has been read using pandas.read_csv with the dataset name "Salary_Data.csv"
```

```python
dataset # After read the csv file dataset 
```

```python
independent=dataset[["YearsExperience"]] # we are splitting the input value of one column  YearsExperience  separately from csvfilename  data by using independent as  variable name 
independent #checking the varaiable 
```

```python
dependent=dataset[["Salary"]] # we are splitting  the input value of one column  salary separately from csvfilename  data by using dependent as  variable name
dependent
```

```python
from sklearn.model_selection import train_test_split # using sklearn we are importing necessary function alone in order to save memory
x_train,x_test,y_train,y_test =train_test_split(independent, dependent ,test_size=0.30,random_state=0)# paramterized function we are giving independent,dependent,test_size,random state value for all these values we are using x_train,x_test,y_train,y_test as variable name
```

```python
from sklearn.linear_model import LinearRegression # from sklearn  contains Ml Alogrithm will import the LinearRegression Algorithm
regressor=LinearRegression() #regressor is a variable name which is  assigned  with a non parametrized function  LinearRegression
regressor.fit(x_train,y_train) # we are calling fit method by giving  the training set input and repective output
```

```python
weight=regressor.coef_ #In order to see the slope value of the regressor we are using weight as a variable name
weight
```

```python
bias=regressor.intercept_ # In order to see the bias value of the regressor we are using bias as a variable name
bias
```

# Evaluation Metrics Process

```python
y_pred=regressor.predict(x_test) # y_predict is a variable evaluation metric should be done for  test dataset which is x_test cross check process
```

```python
from sklearn.metrics import r2_score 
r_score=r2_score(y_test,y_pred) # y_test is a Actual output  and y_pred is the  predicted output
```

```python
r_score # if the r_score value is nearing 1 it is learned well and called as best model  
```

```python
import pickle #  The import statement is used  to save a model
filename="finalized_model_linear.sav" #proper finalized name alone with algorithm name.sav extension
```

```python
pickle.dump(regressor,open(filename,'wb')) # using this dump command we are saving the finalized model
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))# Finally checking once wether it is working as expected
result=loaded_model.predict([[13]])
```
